**Read data**

```python
data = pd.read_csv ('Copynumber_focal_region_input.txt', sep = '\t')
```

**Identify focal gain in Chromosome 10 at specific location. Location is based on gene of interest**

```python
chrom= "chr10"
df=data.loc[(data.Chromosome == chrom)].groupby(by = 'Sample')
```

```python
fg, ax = plt.subplots(figsize = (12,6))
cmap = cm.get_cmap('tab10')
k = -1
for s, d in data.loc[(data.Chromosome == chrom)].groupby(by = 'Sample'):
    k += 1
    for i, r in d.iterrows():
        cn=r["AbsCopy"]
        ax.plot ([r.Start,r.End], [cn,cn], c = cmap(2*k), lw = 3 )
    ax.plot ([],[],c=cmap(2*k), label = s)
ax.set_title (chrom)        
x=80000000
ax.set_xlim (60105460,100105460)
ax.plot ([x,x], [5,0], 'r--', lw = 1) ## x=80000000 line
ax.set_ylim (0,5) ## x=80000000 line y axis co-ordinates
ax.legend(bbox_to_anchor=(1.0, 1))
plt.savefig ("Chr10_focal_gain_at_80000000_position.pdf")
```
